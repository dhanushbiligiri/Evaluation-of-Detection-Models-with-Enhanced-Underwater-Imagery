```python
import zipfile
import os

def zip_folders(folders, zip_filename):
    with zipfile.ZipFile(zip_filename, 'w', zipfile.ZIP_DEFLATED) as zipf:
        for folder in folders:
            for root, dirs, files in os.walk(folder):
                for file in files:
                    file_path = os.path.join(root, file)
                    arcname = os.path.relpath(file_path, folder)
                    zipf.write(file_path, arcname=arcname)

if __name__ == "__main__":
    folders_to_zip = [
        "/kaggle/working/training",  # Replace with the paths to the folders you want to zip

    ]

    output_zip_filename = "output.zip"  # Replace with the desired output ZIP file name

    zip_folders(folders_to_zip, output_zip_filename)
```
